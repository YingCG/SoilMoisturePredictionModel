```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline


mtalbert = pd.read_csv('../Data/soil data/AcrossAucklandRainfallStation/MtAlbert_Data_15mins.csv', parse_dates=['Timestamp (UTC+12:00)'])
mtalbert.head()
```

```python
mtalbert.isna().sum()
```

```python
mtalbert.set_index('Timestamp (UTC+12:00)', inplace=True)
mtalbert_monthly_mean = mtalbert.resample('M').agg({
    'SoilMoisture(%)': 'mean',
    'SoilTemp(c)': 'mean',
    'Rain(mm)': 'sum'
})
print("Mt Albert Data Monthly Average")
print(mtalbert_monthly_mean)
```

```python
nan_rows = mtalbert_monthly_mean[mtalbert_monthly_mean.isna().any(axis=1)]
print(nan_rows)
```

```python
ararimu = pd.read_csv('../Data/soil data/AcrossAucklandRainfallStation/ArarimuCombined_Data_15mins.csv', parse_dates=['Timestamp (UTC+12:00)'])
ararimu.head()
```

```python
ararimu.set_index('Timestamp (UTC+12:00)', inplace=True)
ararimu_monthly_mean = mtalbert.resample('M').agg({
    'SoilMoisture(%)': 'mean',
    'SoilTemp(c)': 'mean',
    'Rain(mm)': 'sum'
})
print("Ararimu Data Monthly Average")
print(ararimu_monthly_mean)
```

```python
nan_rows = ararimu_monthly_mean[ararimu_monthly_mean.isna().any(axis=1)]
print(nan_rows)
```

```python
print(ararimu_monthly_mean.index)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Filter data from May 2019 onwards
ararimu_from_may19 = ararimu_monthly_mean.loc['2019-05-31':]

plt.figure(figsize=(10, 5))

# Plot Soil Moisture
plt.subplot(3, 1, 1)
plt.plot(ararimu_from_may19.index, ararimu_from_may19['SoilMoisture(%)'], marker='o', label='Soil Moisture (%)', color='b')
plt.title('Soil Moisture Over Time (From May 2019)')
plt.ylabel('Soil Moisture (%)')
plt.xticks(rotation=45)
plt.grid()
# plt.legend()

# Plot Soil Temperature
plt.subplot(3, 1, 2)
plt.plot(ararimu_from_may19.index, ararimu_from_may19['SoilTemp(c)'], marker='o', label='Soil Temperature (°C)', color='r')
plt.title('Soil Temperature Over Time (From May 2019)')
plt.ylabel('Soil Temperature (°C)')
plt.xticks(rotation=45)
plt.grid()
# plt.legend()

# Plot Rainfall
plt.subplot(3, 1, 3)
plt.plot(ararimu_from_may19.index, ararimu_from_may19['Rain(mm)'], marker='o', label='Rainfall (mm)', color='g')
plt.title('Rainfall Over Time (From May 2019)')
plt.ylabel('Rainfall (mm)')
plt.xticks(rotation=45)
plt.grid()
# plt.legend()

plt.tight_layout()
plt.show()
```
